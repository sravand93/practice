### Performing PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
num_data = scale(num_data)
```

```python
pca_for_data = PCA(n_components=0.95,svd_solver="full")
```

```python
pca_for_data.fit(num_data)
```

```python
pca_for_data.explained_variance_ratio_
```

```python
pca_for_data.explained_variance_
```

```python
import numpy as np
```

```python
#explains cummulative variance
var_pca = np.cumsum(np.round(pca_for_data.explained_variance_ratio_,decimals=4)*100)
```

```python
print(var_pca)
```

```python
plt.plot(var_pca)
```

```python
pca_data = pca_for_data.transform(num_data)
```

```python
X = diagnosis_data.iloc[:,:-1]
Y = diagnosis_data.V49
```

## Test,train and validation split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.20,random_state = 102)
```

```python
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train,test_size = 0.20,random_state=102)
```

```python
x_train.shape,x_test.shape,x_val.shape
```

```python
#standardizing the numeric values

from sklearn import preprocessing
std = preprocessing.StandardScaler()
std.fit(x_train)
x_train = std.transform(x_train)
x_test = std.transform(x_test)
x_val = std.transform(x_val)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
logistic_preds =log_model.predict(x_test)
```

### Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,logistic_preds))
```

```python
from sklearn import metrics
```

```python
print("Accuracy", metrics.accuracy_score(y_test,logistic_preds))
print("Recall  ", metrics.recall_score(y_test,logistic_preds,average="macro"))
print("precision",metrics.precision_score(y_test,logistic_preds,average="macro"))
```
