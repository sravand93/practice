```python
import pandas as pd
import numpy as np
```

```python
# importing the movie lens data

r_cols = ['user_id','movie_id','ratings']
ratings = pd.read_csv('E:\\Udemy\\sundog_Datascience\\DataScience-Python3\\ml-100k\\u.data',sep='\t',names=r_cols,usecols=range(3),encoding="ISO-8859-1")

m_cols = ['movie_id','title']
movies = pd.read_csv('E:\\Udemy\\sundog_Datascience\\DataScience-Python3\\ml-100k\\u.item',sep='|', names=m_cols,usecols=range(2),encoding="ISO-8859-1")
```

```python
#merging both the data

ratings = pd.merge(movies,ratings)
ratings.head()
```

```python
#creating a user/movie rating matrix using pivot_table
movie_ratings = ratings.pivot_table(index=['user_id'],columns=['title'],values='ratings')
movie_ratings.head()
#All the NAN values indicate the missing values i.e., the user has not watched that movie yet
```

```python
#now we will do item based filtering 
starwar_ratings = movie_ratings['Star Wars (1977)']
starwar_ratings.head()
```

```python
similar_movies = movie_ratings.corrwith(starwar_ratings)
```

```python
similar_movies=similar_movies.dropna()
df = pd.DataFrame(similar_movies)
```

```python
df.head(10)
```

```python
similar_movies.sort_values(ascending=False)
```

```python
movieStats = ratings.groupby('title').agg({'ratings': [np.size,np.mean]})
movieStats.head(10)
```

```python
popular_movies = movieStats['ratings']['size']>= 200
movieStats[popular_movies].sort_values([('ratings','mean')],ascending = False)[:15]
```

```python
df = movieStats[popular_movies].join(pd.DataFrame(similar_movies,columns=['similarity']))
```
