```python
#reading the data
data = pd.read_excel("Aeron_Numbers_INSOFE.xlsx")
```

```python
#removing unnamed column
data.drop(['Unnamed: 1'],axis=1,inplace = True)
```

```python
#now drop the above rows using index format
data.drop(data.index[[0,1,12,25,38,51,64,77,90,103,116,129,142,155]],inplace=True)
```

```python
data.head()
```

```python
#convrting the object to date format

data['Month/Year'] = pd.to_datetime(data['Month/Year'],format ='%b-%Y')
```

```python
data['Year'] = pd.DatetimeIndex(data['Month/Year']).year
data['month'] = pd.DatetimeIndex(data['Month/Year']).month
```

```python
date = data['Month/Year']
```

```python
data.drop('Month/Year',axis = 1,inplace = True)
```

```python
#replacing all the - with 0s
data.replace(to_replace= "–",value= 0,inplace= True)
```

```python
data
```

```python
data = data.apply(lambda x: pd.to_numeric(x.astype(str).str.replace(",",""),errors="coerce"))
```

```python
data.head(10)
```

```python
data.columns
```

```python
#renaming the columns
data.rename(columns={'CAT 0': 'Transactions_0',     'Unnamed: 3':'total_0',
                     'CAT 1': 'Transactions_1',     'Unnamed: 5':'total_1',
                     'CAT 2': 'Transactions_2',     'Unnamed: 7':'total_2',
                     'CAT 2.1': 'Transactions_2.1', 'Unnamed: 9':'total_2.1',
                     'CAT 2.2': 'Transactions_2.2', 'Unnamed: 11':'total_2.2',
                     'CAT 2.3': 'Transactions_2.3', 'Unnamed: 13':'total_2.3',
                     'CAT 2.4': 'Transactions_2.4', 'Unnamed: 15':'total_2.4',
                     'CAT 2.5': 'Transactions_2.5', 'Unnamed: 17':'total_2.5',
                     'CAT 3'  : 'Transactions_3',   'Unnamed: 19':'total_3',
                     'CAT 3.1': 'Transactions_3.1', 'Unnamed: 21':'total_3.1',
                     'CAT 3.2': 'Transactions_3.2', 'Unnamed: 23':'total_3.2'}, inplace=True)
```

```python
data.head()
```

```python
#converting the number of transactions to integer from float as they should not be in decimal 
data['Transactions_0']= round(data['Transactions_0']).astype('int')
data['Transactions_1'] = round(data['Transactions_1']).astype('int')
data['Transactions_2'] = round(data['Transactions_2']).astype('int')
data['Transactions_2.1'] = round(data['Transactions_2.1']).astype('int')
data['Transactions_2.2'] = round(data['Transactions_2.2']).astype('int')
data['Transactions_2.3'] = round(data['Transactions_2.3']).astype('int')
data['Transactions_2.4'] = round(data['Transactions_2.4']).astype('int')
data['Transactions_2.5'] = round(data['Transactions_2.5']).astype('int')
data['Transactions_3'] = round(data['Transactions_3']).astype('int')
data['Transactions_3.1'] = round(data['Transactions_3.1']).astype('int')
data['Transactions_3.2'] = round(data['Transactions_3.2']).astype('int')
```
